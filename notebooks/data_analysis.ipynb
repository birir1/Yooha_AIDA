```python
# ============================================================
# REAL DATASET DOWNLOADER
# Memory-Augmented AI Empathetic Assistant
# ============================================================

import os
import requests
import zipfile
import tarfile

# ============================================================
# ROOT PATHS
# ============================================================

PROJECT_ROOT = "/root/workspace/memory_augmented_ai_assistant"

RAW_CONV = f"{PROJECT_ROOT}/data/raw/conversations"
RAW_EMO = f"{PROJECT_ROOT}/data/raw/emotional_datasets"
RAW_KNOW = f"{PROJECT_ROOT}/data/raw/external_knowledge"
EVAL_DIR = f"{PROJECT_ROOT}/data/evaluation/benchmark_sets"

os.makedirs(RAW_CONV, exist_ok=True)
os.makedirs(RAW_EMO, exist_ok=True)
os.makedirs(RAW_KNOW, exist_ok=True)
os.makedirs(EVAL_DIR, exist_ok=True)

# ============================================================
# DOWNLOAD FUNCTION
# ============================================================

def download_file(url, save_path):

    print(f"\nDownloading:\n{url}")

    response = requests.get(url, stream=True)

    response.raise_for_status()

    with open(save_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)

    print(f"Saved -> {save_path}")

# ============================================================
# EXTRACT ZIP
# ============================================================

def extract_zip(zip_path, extract_to):

    print(f"\nExtracting: {zip_path}")

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

    print(f"Extracted -> {extract_to}")

# ============================================================
# EXTRACT TAR.GZ
# ============================================================

def extract_tar(tar_path, extract_to):

    print(f"\nExtracting: {tar_path}")

    with tarfile.open(tar_path, 'r:gz') as tar_ref:
        tar_ref.extractall(extract_to)

    print(f"Extracted -> {extract_to}")

# ============================================================
# 1. GOEMOTIONS
# REAL EMOTION DATASET
# ============================================================

go_dir = f"{RAW_EMO}/go_emotions"
os.makedirs(go_dir, exist_ok=True)

go_url = (
    "https://storage.googleapis.com/"
    "gresearch/goemotions/data/full_dataset/goemotions_1.csv"
)

download_file(
    go_url,
    f"{go_dir}/goemotions_1.csv"
)

# ============================================================
# 2. DAILYDIALOG
# REAL MULTI-TURN CONVERSATION DATASET
# ============================================================

daily_dir = f"{RAW_CONV}/daily_dialog"
os.makedirs(daily_dir, exist_ok=True)

daily_url = (
    "http://yanran.li/files/ijcnlp_dailydialog.zip"
)

daily_zip = f"{daily_dir}/daily_dialog.zip"

download_file(
    daily_url,
    daily_zip
)

extract_zip(
    daily_zip,
    daily_dir
)

# ============================================================
# 3. EMPATHETIC DIALOGUES
# REAL EMPATHY DATASET
# ============================================================

empathetic_dir = f"{RAW_CONV}/empathetic_dialogues"
os.makedirs(empathetic_dir, exist_ok=True)

empathetic_url = (
    "https://dl.fbaipublicfiles.com/parlai/empatheticdialogues/"
    "empatheticdialogues.tar.gz"
)

empathetic_tar = f"{empathetic_dir}/empatheticdialogues.tar.gz"

download_file(
    empathetic_url,
    empathetic_tar
)

extract_tar(
    empathetic_tar,
    empathetic_dir
)

# ============================================================
# 4. PERSONA CHAT
# REAL MEMORY/PERSONA DATASET
# ============================================================

persona_dir = f"{RAW_CONV}/persona_chat"
os.makedirs(persona_dir, exist_ok=True)

persona_url = (
    "https://s3.amazonaws.com/datasets.huggingface.co/"
    "personachat/personachat_self_original.json"
)

download_file(
    persona_url,
    f"{persona_dir}/personachat_self_original.json"
)

# ============================================================
# 5. SQUAD
# FACTUAL QA DATASET
# ============================================================

squad_dir = f"{RAW_KNOW}/squad"
os.makedirs(squad_dir, exist_ok=True)

download_file(
    "https://rajpurkar.github.io/SQuAD-explorer/dataset/train-v2.0.json",
    f"{squad_dir}/train-v2.0.json"
)

download_file(
    "https://rajpurkar.github.io/SQuAD-explorer/dataset/dev-v2.0.json",
    f"{squad_dir}/dev-v2.0.json"
)

# ============================================================
# 6. WIKITEXT-2
# LONG CONTEXT MEMORY DATASET
# ============================================================

wiki_dir = f"{RAW_KNOW}/wikitext"
os.makedirs(wiki_dir, exist_ok=True)

wiki_url = (
    "https://s3.amazonaws.com/research.metamind.io/wikitext/"
    "wikitext-2-v1.zip"
)

wiki_zip = f"{wiki_dir}/wikitext-2-v1.zip"

download_file(
    wiki_url,
    wiki_zip
)

extract_zip(
    wiki_zip,
    wiki_dir
)

# ============================================================
# 7. FEVER
# FACT VERIFICATION / HALLUCINATION
# ============================================================

fever_dir = f"{EVAL_DIR}/fever"
os.makedirs(fever_dir, exist_ok=True)

fever_url = (
    "https://fever.ai/download/fever/train.jsonl"
)

download_file(
    fever_url,
    f"{fever_dir}/train.jsonl"
)

# ============================================================
# COMPLETE
# ============================================================

print("\n" + "="*60)
print("REAL DATASETS DOWNLOADED SUCCESSFULLY")
print("="*60)
```

```python
# ============================================================
# PERSONACHAT PREPROCESSING PIPELINE
# Builds REAL datasets for the capstone
# ============================================================

import os
import pandas as pd

PROJECT_ROOT = "/root/workspace/memory_augmented_ai_assistant"

PERSONA_FILE = (
    f"{PROJECT_ROOT}/data/raw/conversations/persona_chat/"
    "train_self_original.txt"
)

# ============================================================
# OUTPUT FILES
# ============================================================

MEMORY_OUT = (
    f"{PROJECT_ROOT}/data/raw/conversations/"
    "memory_training_dataset/train.csv"
)

MULTI_OUT = (
    f"{PROJECT_ROOT}/data/raw/conversations/"
    "multi_turn_dialogues/dialogues.csv"
)

EMPATHY_OUT = (
    f"{PROJECT_ROOT}/data/raw/conversations/"
    "synthetic_empathy_dataset/train.csv"
)

LONG_MEMORY_OUT = (
    f"{PROJECT_ROOT}/data/raw/memory_logs/"
    "long_term_memory/memory_dataset.csv"
)

# ============================================================
# CREATE DIRECTORIES
# ============================================================

for path in [
    MEMORY_OUT,
    MULTI_OUT,
    EMPATHY_OUT,
    LONG_MEMORY_OUT
]:
    os.makedirs(os.path.dirname(path), exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

print("Loading PersonaChat dataset...")

with open(PERSONA_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

print(f"Loaded {len(lines)} lines")

# ============================================================
# DATA CONTAINERS
# ============================================================

memory_rows = []
dialogue_rows = []
empathy_rows = []
long_memory_rows = []

conversation_id = 0

persona_memories = []

# ============================================================
# PROCESS DATA
# ============================================================

for line in lines:

    line = line.strip()

    if not line:
        continue

    try:

        parts = line.split(" ", 1)

        if len(parts) < 2:
            continue

        content = parts[1]

        # ====================================================
        # PERSONA MEMORY
        # ====================================================

        if content.startswith("your persona:"):

            persona_text = (
                content.replace("your persona:", "")
                .strip()
            )

            persona_memories.append(persona_text)

            long_memory_rows.append({
                "memory": persona_text,
                "memory_type": "persona_profile",
                "importance_score": 0.95
            })

        # ====================================================
        # DIALOGUES
        # ====================================================

        elif "\t" in content:

            sections = content.split("\t")

            if len(sections) >= 2:

                user_msg = sections[0].strip()
                assistant_msg = sections[1].strip()

                # ============================================
                # MEMORY TRAINING
                # ============================================

                memory_rows.append({
                    "previous_context": user_msg,
                    "new_message": assistant_msg,
                    "memory_type": "dialogue_memory"
                })

                # ============================================
                # MULTI TURN
                # ============================================

                dialogue_rows.append({
                    "conversation_id": conversation_id,
                    "turn_id": 0,
                    "speaker": "user",
                    "message": user_msg
                })

                dialogue_rows.append({
                    "conversation_id": conversation_id,
                    "turn_id": 1,
                    "speaker": "assistant",
                    "message": assistant_msg
                })

                # ============================================
                # EMPATHY DATASET
                # ============================================

                empathy_rows.append({
                    "user_input": user_msg,
                    "assistant_response": assistant_msg
                })

                conversation_id += 1

    except Exception as e:
        continue

# ============================================================
# SAVE DATASETS
# ============================================================

memory_df = pd.DataFrame(memory_rows)
memory_df.to_csv(MEMORY_OUT, index=False)

print(f"Saved -> {MEMORY_OUT}")

dialogue_df = pd.DataFrame(dialogue_rows)
dialogue_df.to_csv(MULTI_OUT, index=False)

print(f"Saved -> {MULTI_OUT}")

empathy_df = pd.DataFrame(empathy_rows)
empathy_df.to_csv(EMPATHY_OUT, index=False)

print(f"Saved -> {EMPATHY_OUT}")

long_memory_df = pd.DataFrame(long_memory_rows)
long_memory_df.to_csv(LONG_MEMORY_OUT, index=False)

print(f"Saved -> {LONG_MEMORY_OUT}")

# ============================================================
# SUMMARY
# ============================================================

print("\n" + "=" * 60)

print(f"Memory Samples: {len(memory_df)}")
print(f"Dialogue Turns: {len(dialogue_df)}")
print(f"Empathy Samples: {len(empathy_df)}")
print(f"Long-Term Memories: {len(long_memory_df)}")

print("=" * 60)
```
